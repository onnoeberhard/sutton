```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize, ListedColormap
from oetils.plotting import init_plotting
from tqdm.notebook import trange
```

```python
W = init_plotting()
```

The respawning in the original description is horrible for MC rollouts, because they can become very, very long. I have changed the dynamics so that hitting a wall will stop the agent and incur a strong negative reward.

```python
# Build race track
track = np.zeros((30, 30), int)
track[:, 5:15] = 1
track[-5:, 10:] = 1
track[0, 5:16] = 2
track[-5:, -1] = 3
for i in range(11):
    track[14 + i, 15 + i] = 1
    track[15 + i, 15 + i] = 1
    track[16 + i, 15 + i] = 1

track[1, 15] = 1
for i in range(13):
    for j in range(i, i + 4):
        track[1 + i + j, 16 + i] = 1

cm = ListedColormap(['white', 'lightgrey', 'green', 'red'])

fig, ax = plt.subplots()
ax.imshow(track, cmap=cm)
ax.invert_yaxis()
```

```python
def mdp(s, a, rng):
    """MDP dynamics"""
    x, y, vx, vy = s
    ax, ay = a
    
    illegal = vx + ax < 0 or vy + ay < 0
    
    vx = max(min(vx + ax, 5), 0)
    vy = max(min(vy + ay, 5), 0)
    x_ = x + vx
    y_ = y + vy
    
    
    if illegal or any(int(x + i/5*vx) < track.shape[1] and int(y + i/5*vy) < track.shape[0] and 
          track[int(y + i/5*vy), int(x + i/5*vx)] == 0 for i in range(6)):
        return (x, y, 0, 0), -10

    if any(int(x + i/5*vx) < track.shape[1] and int(y + i/5*vy) < track.shape[0] 
           and track[int(y + i/5*vy), int(x + i/5*vx)] == 3 for i in range(6)):
        return (x_, y_, vx, vy), 0

    if x_ >= track.shape[1] or y_ >= track.shape[0]:
        return (x, y, 0, 0), -10
        
    return (x_, y_, vx, vy), -1
```

```python
# On-policy Monte Carlo control (epsilon-greedy)
rng = np.random.default_rng(42)

eps = 0.3

Q = np.full((*track.shape, 6, 6, 9), np.nan)
N = np.zeros((*track.shape, 6, 6, 9))
pi = np.full((*track.shape, 6, 6, 9), 1/9)

y0s, x0s = np.where(track == 2)
for i in trange(50000):
    # Generate episode
    start = rng.choice(len(x0s))
    s = x0s[start], y0s[start], 0, 0
    ep = []
    while True:
        a = np.unravel_index(np.argmax(rng.multinomial(1, pi[s])), (3, 3))
        a = a[0] - 1, a[1] - 1
        s_, r = mdp(s, a, rng)
        ep.append((s, a, s_, r))
        if r == 0: break
        s = s_
    
    # Update value function and policy
    G = 0
    for t in range(len(ep) - 1, -1, -1):
        s, a, s_, r = ep[t]
        G += r
        if any((s == s1 and a == a1) for (s1, a1, _, _) in ep[:t]):
            continue
        a = np.ravel_multi_index((a[0] + 1, a[1] + 1), (3, 3))
        N[s][a] += 1
        if np.isnan(Q[s][a]):
            Q[s][a] = 0
        Q[s][a] += 1/N[s][a] * (G - Q[s][a])
        pi[s] = eps/9
        pi[s][np.argmax(Q[s])] += 1 - eps
```

```python
# Plot solutions
fig, ax = plt.subplots()
ax.imshow(track, cmap=cm)
ax.invert_yaxis()
xl, yl = ax.get_xlim(), ax.get_ylim()

Rs = []
for start in range(len(x0s)):
    R = 0
    s = x0s[start], y0s[start], 0, 0
    ep = []
    while True:
        a = np.unravel_index(np.argmax(pi[s]), (3, 3))
        a = a[0] - 1, a[1] - 1
        s_, r = mdp(s, a, rng)
        R += r
        ep.append((s, a, s_, r))
        if r == 0: break
        s = s_
        
    Rs.append(R)
    xs = [s[0] for (s, a, s_, r) in ep] + [ep[-1][2][0]]
    ys = [s[1] for (s, a, s_, r) in ep] + [ep[-1][2][1]]
    ax.plot(xs, ys, '.-')
    ax.set_xlim(xl)
    ax.set_ylim(yl)

print(Rs)
```

```python
# Plot value function
V = (Q * pi).sum(-1)
norm = Normalize(np.nanmin(V), np.nanmax(V))

fig, ax = plt.subplots(6, 6, sharex=True, sharey=True, figsize=(1.1*W, W))
for i in range(6):
    for j in range(6):
        ax[i, j].imshow(V[..., i, j].T, norm=norm)
        ax[i, j].invert_yaxis()

fig.colorbar(ScalarMappable(norm=norm), ax=ax[:, -1], aspect=50, pad=0.1)
```
