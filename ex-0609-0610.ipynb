# Original problem

```python
def mdp(s, a):
    # Windy gridworld MDP
    x, y = s

    # Wind
    if x in [3, 4, 5, 8]:
        y += 1
    if x in [6, 7]:
        y += 2

    # Actions
    if a == 0:
        y += 1
    elif a == 1:
        y -= 1
    elif a == 2:
        x += 1
    elif a == 3:
        x -= 1

    # Clip
    x = min(max(x, 0), 9)
    y = min(max(y, 0), 6)
    
    r = 0 if (x, y) == (7, 3) else -1
    return (x, y), r
```

```python
def sarsa(rng, alpha=0.1, eps=0.1):
    Q = np.zeros((10, 7, 4))
    for i in trange(10000):
        s = (0, 3)
        a = np.argmax(Q[s]) if rng.uniform() > eps else rng.integers(4)
        while True:
            s_, r = mdp(s, a)
            a_ = np.argmax(Q[s_]) if rng.uniform() > eps else rng.integers(4)
            Q[s][a] += alpha * (r + Q[s_][a_] - Q[s][a])
            s, a = s_, a_
            if r == 0: break
    return Q
```

```python
rng = np.random.default_rng(42)
Q = sarsa(rng)
```

```python
fig, ax = plt.subplots()
I = ax.pcolormesh(np.arange(10), np.arange(7), Q.max(-1).T)
fig.colorbar(I, ax=ax, pad=0.01, label="$V$")

ep = [(0, 3)]
s = ep[-1]
while True:
    a = np.argmax(Q[s])
    s, r = mdp(s, a)
    ep.append(s)
    if r == 0: break

plt.plot(*np.asarray(ep).T, 'r.-')
```

# Stochastic Wind

```python
def mdp(s, a, rng):
    # Windy gridworld MDP
    x, y = s

    # Wind
    if x in [3, 4, 5, 8]:
        y += rng.integers(0, 3)
    if x in [6, 7]:
        y += rng.integers(1, 4)

    # Actions
    if a == 0 or a == 4 or a == 7:
        y += 1
    if a == 1 or a == 5 or a == 6:
        y -= 1
    if a == 2 or a == 4 or a == 5:
        x += 1
    if a == 3 or a == 6 or a == 7:
        x -= 1

    # Clip
    x = min(max(x, 0), 9)
    y = min(max(y, 0), 6)
    
    r = 0 if (x, y) == (7, 3) else -1
    return (x, y), r
```

```python
def sarsa(rng, alpha=0.1, eps=0.1):
    Q = np.zeros((10, 7, 8))
    for i in trange(10000):
        s = (0, 3)
        a = np.argmax(Q[s]) if rng.uniform() > eps else rng.integers(8)
        while True:
            s_, r = mdp(s, a, rng)
            a_ = np.argmax(Q[s_]) if rng.uniform() > eps else rng.integers(8)
            Q[s][a] += alpha * (r + Q[s_][a_] - Q[s][a])
            s, a = s_, a_
            if r == 0: break
    return Q
```

```python
rng = np.random.default_rng(42)
Q = sarsa(rng)
```
